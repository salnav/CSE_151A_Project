# Data numeric exploration

```python
import pandas as pd
import os
```

```python
folder_path = '/content/drive/MyDrive/character+font+images'

csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

dataframes = [pd.read_csv(os.path.join(folder_path, file)) for file in csv_files]

df = pd.concat(dataframes, ignore_index=True)

df.to_csv('character+font+images.csv', index=False)

df
```

```python
df = df.sample(frac=0.4, random_state=5)
```

```python
df.head()
```

```python
df.describe()
```

```python
print(f"Number of observations (images): {df.shape[0]}")
```

```python
print(df.columns)
```

```python
num_classes = df['m_label'].nunique()
print(f'Number of unique classes (characters): {num_classes}')
print(f'Min class label: {df["m_label"].min()}')
print(f'Max class label: {df["m_label"].max()}')
```

```python
font_counts = df['font'].value_counts()
print('Font counts:')
print(font_counts)
```

```python
total_missing_data = df[df.isnull().any(axis=1)].shape[0]

print(f'Number of rows with missing data: {total_missing_data}')

missing_data_per_column = df.isnull().sum()

print('Missing data per column:')
print(missing_data_per_column)
```

## Data Visual Exploration

```python
import matplotlib.pyplot as plt
```

```python
examples = 0
for index, row in df.iterrows():

    label = row['m_label']
    h = row['originalH']
    w = row['originalW']


    pixel_values = row.iloc[12:412].values.astype(int).reshape((20, 20))

    plt.imshow(pixel_values, cmap='gray', vmin=0, vmax=255)
    plt.title(f'Label: {chr(label)}, Original Height: {h}, Original Width: {w}')
    plt.colorbar()
    plt.show()
    if examples >= 15:
        break
    examples += 1
```
